Nullity correlation ranges from -1 (if one variable appears the other definitely does not) to 0 (variables appearing or not appearing have no effect on one another) to 1 (if one variable appears the other definitely also does).

```python
#fill up manage_shop_indicator with 0
data["manage_shop_indicator"].fillna(0,inplace=True)
```

## Dataset splitting

```python
whitelist = data[data['decorated_indicator'] == 1]
non_whitelist = data[data['decorated_indicator'] == 0]
```

Pre-processing for whitelisted shop data
- null value

```python
msno.bar(whitelist)
```

```python
# for variables with few missing values, drop them (as the no. of missing values is vary small)
whitelist.dropna(subset=["shop_history_order"],inplace=True)
whitelist.dropna(subset=["weighted_shop_rating"],inplace=True)
```

```python
whitelist.info()
```

```python
whitelist_with_null = whitelist.values
ix = [i for i in range(whitelist_with_null.shape[1]) if i != 4]
X, y = whitelist_with_null[:, ix], whitelist_with_null[:, 4]
# evaluate each strategy on the dataset
results = list()
strategies = [str(i) for i in [15]]#7,9,15,18,20
for s in strategies:
    # create the modeling pipeline
    pipeline = Pipeline(steps=[('i', KNNImputer(n_neighbors=int(s))), ('m', RandomForestClassifier())])
    # evaluate the model
    cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
    scores = cross_val_score(pipeline, X, y, scoring='accuracy', cv=cv, n_jobs=-1)
    # store results
    results.append(scores)
    print('>%s %.3f (%.3f)' % (s, mean(scores), std(scores)))
```

```python
#K-NN null value fill-up
imputer = KNNImputer(n_neighbors=18, weights='uniform', metric='nan_euclidean')
imputer.fit(whitelist)
whitelist_filled = imputer.transform(whitelist)
```

```python
whitelist_filled=pd.DataFrame(whitelist_filled,columns=list(data.columns))
```
